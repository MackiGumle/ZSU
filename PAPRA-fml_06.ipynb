```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Elbow Method
sse = []
silhouette_scores = []
k_range = range(2, 10)  # Test from 2 to 9 clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_min_max_scaled)
    sse.append(kmeans.inertia_)  # Sum of squared distances
    silhouette_scores.append(silhouette_score(X_min_max_scaled, kmeans.labels_))

# Plotting the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
ax1.plot(k_range, sse, marker='o')
ax1.set_title("Elbow Method")
ax1.set_xlabel("Number of clusters")
ax1.set_ylabel("Sum of Squared Distances")


```

```python

optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_min_max_scaled)

df_scaled['Cluster'] = clusters
df['Cluster'] = clusters
```

```python
# Plot feature distributions per cluster
df['Cluster'] = kmeans.fit_predict(X_min_max_scaled)
sns.pairplot(df, hue='Cluster')
plt.show()
```
